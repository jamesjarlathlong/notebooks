```python
import pandas as pd
import seaborn as sns
sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
})
import matplotlib.pyplot as plt
import itertools
%matplotlib inline
df = pd.read_csv('total8fdd', names = ['id','t (s)','p','reduce node'])
```

```python
import matplotlib.text as mpl_text

class AnyObject(object):
    def __init__(self, text, color):
        self.my_text = text
        self.my_color = color

class AnyObjectHandler(object):
    def legend_artist(self, legend, orig_handle, fontsize, handlebox):
        print(orig_handle)
        x0, y0 = handlebox.xdescent, handlebox.ydescent
        width, height = handlebox.width, handlebox.height
        patch = mpl_text.Text(x=7.5, y=0, text=orig_handle.my_text, color=orig_handle.my_color, verticalalignment=u'baseline', 
                                horizontalalignment=u'left', multialignment=None, 
                                fontproperties=None, rotation=0, linespacing=None, 
                                rotation_mode=None)
        handlebox.add_artist(patch)
        return patch
```

```python
fig,ax = plt.subplots(1)
#fig.set_size_inches(10, 4)
fig.set_size_inches(6,4)
def t(x):
    if x=='0':
        return 'sink'
    if x=='21':
        return 'optimal'
    if x=='41':
        return 'worst'
ax = sns.stripplot(x="reduce node", y="t (s)", data=df,
                   jitter=0.075,palette=['#252525','#a50f15','#737373']
                  ,label='observed', alpha=0.6)
#sns.boxplot(x="s", y="t", data=df, palette=['#f05f40','#252525']
#                   ,hue='strategy')
# distance across the "X" or "Y" stipplot column to span, in this case 40%
median_width = 0.4
mu = u'\u03BC'
rho=u'\u03C1'
for tick, text in zip(ax.get_xticks(), ax.get_xticklabels()):
    sample_name = text.get_text()  # "X" or "Y"
    print('sample name: ', sample_name)
    # calculate the median value for all replicates of either X or Y
    median_val1 = df[df['reduce node']==int(sample_name)]['p'].median()
    plotter = median_val1-15
    median_obs = df[df['reduce node']==int(sample_name)]['t (s)'].mean()
    print('plotter; ',median_val1)
    # plot horizontal lines across the column, centered on the tick
    if tick == 0:
        ax.plot([tick-median_width/2, tick+median_width/2], [median_val1, median_val1],
                lw=1.25, color='#252525',label='predicted')
        ax.text(tick+median_width/2+0.05, median_val1, '{}={}'.format(rho, round(median_val1/1000,1)))
        ax.text(tick+median_width/2+0.05, median_val1-5000, '{}={}'.format(mu, round(median_obs/1000,1)))
    else:
        ax.plot([tick-median_width/2, tick+median_width/2], [median_val1, median_val1],
                lw=1.25, color='#252525')
        ax.text(tick+median_width/2+0.05, median_val1, '{}={}'.format(rho, round(median_val1/1000,1)))
        ax.text(tick+median_width/2+0.05,median_val1-5000, '{}={}'.format(mu, round(median_obs/1000,1)))
        
ylabels = [int(label/1000) for label in ax.get_yticks()]
ax.set_yticklabels(ylabels)
handles, labels = ax.get_legend_handles_labels()
mu_obj = AnyObject(mu, "#252525")
rho_obj = AnyObject(rho, "#252525")
legend = ax.legend(handles[0:2]+[rho_obj, mu_obj], labels[0:2]+['predicted value','mean observed value'],
                   handler_map={rho_obj:AnyObjectHandler(), mu_obj:AnyObjectHandler()}
                   ,loc='best')
#legend = ax.legend(loc='best')
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('white')
sns.despine()
fig.tight_layout()
fig.savefig('8resfdd.png', format='png', dpi=600)
```

```python
import math
rssi = {96: {0: -83, 53: -23, 55: -36, 41: -83, 95: -33, 63: -28, 21: -46, 15: -34}, 53: {0: -74, 96: -23, 21: -34, 55: -27, 63: -25, 95: -29, 15: -31}, 15: {0: -75, 96: -34, 53: -31, 55: -57, 95: -34, 21: -42, 63: -37}, 41: {96: -83, 0: -89}, 63: {0: -81, 96: -28, 53: -25, 55: -28, 15: -37, 21: -38, 95: -35}, 21: {0: -77, 96: -46, 53: -34, 55: -51, 15: -42, 63: -38}, 95: {96: -33, 55: -37, 63: -35, 53: -29, 15: -34}}
posit = lambda d:{target:-1*value for target, value in d.items()}
pos_rssi = {k: posit(v) for k,v in rssi.items()}
tplify = lambda k,vals: [(k,v,rssi) for v, rssi in vals.items()]
wrapup = lambda triplelist: {(item[0], item[1]):item[2] for item in triplelist}
flattener = lambda lsts: itertools.chain(*lsts)
tplified = list(flattener([tplify(source, targets) for source, targets in pos_rssi.items()]))
edge_labels = wrapup(tplified)
weighter = lambda d: {k:{'weight':v} for k,v in d.items()}
weighted_rssi = {k:weighter(v) for k,v in pos_rssi.items()}
import networkx as nx
fig, ax = plt.subplots(1)
fig.set_size_inches(6,6)
g = nx.DiGraph()
g.add_weighted_edges_from(tplified)
n = len(rssi)
init = {k:(k[0]/96,k[1]/96) for k,v in edge_labels.items()}
pos=nx.spring_layout(g, pos=init,fixed=[0],weight='weight',k=0.5, iterations=1000)
def sample_labels(edge_labels):
    return{k:v for k,v in edge_labels.items() if k[0] in [0,41,21] or k[1] in [0,41,21]}
#print('g: ',g)
colors = ['#252525' for i in range(9)]
print(tplified)
node_idxs = {k:idx for idx,k in enumerate(g)}
print(node_idxs)
colors[node_idxs[21]]='#a50f15'
colors[node_idxs[41]] ='#737373'
bprops = dict(boxstyle="circle,pad=0", fc='white', ec='white',alpha = 1, lw=0)
a = nx.draw(g,pos,ax=ax,with_labels=False, node_color =colors, alpha=0.95,arrows=False)
a = nx.draw_networkx_labels(g,pos,font_size=10,  font_color='#d9d9d9')
a = nx.draw_networkx_edge_labels(g, pos, edge_labels=sample_labels(edge_labels), label_pos=0.5
                             ,font_size=8, font_color='#525252', font_family='serif'
                            ,font_weight='normal', alpha=1, bbox=bprops
                             ,ax=ax, rotate=True)
fig.savefig('graph.png', dpi=600, format='png')
```

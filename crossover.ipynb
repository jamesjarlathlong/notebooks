```python
import pandas as pd
import seaborn as sns
sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
})
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('totals', names = ['id','t','p','m','px','s'])
df
```

```python
fig,ax = plt.subplots(1)
#fig.set_size_inches(10, 4)
df['strategy'] = df['m'].apply(lambda x: 'centralised' if x=='[0]' else 'decentralised')
fig.set_size_inches(6,4)
#ax = sns.stripplot(x="s", y="t", data=df, palette=['#f05f40','#252525']
#                   ,hue='strategy',dodge=0.5, s=5, alpha=0.6,size=4)
ax = sns.stripplot(x="s", y="t", data=df, palette=['#f05f40','#252525']
                   ,hue='strategy',dodge=True)
#sns.boxplot(x="s", y="t", data=df, palette=['#f05f40','#252525']
#                   ,hue='strategy')
# distance across the "X" or "Y" stipplot column to span, in this case 40%
median_width = 0.5


for tick, text in zip(ax.get_xticks(), ax.get_xticklabels()):
    sample_name = text.get_text()  # "X" or "Y"
    print('sample name: ', sample_name)
    # calculate the median value for all replicates of either X or Y
    matching = df[df['s']==float(sample_name)]
    median_val1 = matching[matching['strategy']=='decentralised']['p'].median()
    median_val2 = matching[matching['strategy']=='centralised']['p'].median()
    print(median_val1)
    # plot horizontal lines across the column, centered on the tick
    if tick == 0:
        ax.plot([tick-median_width/2, tick], [median_val1, median_val1],
                lw=1, color='k',label='predicted')
        ax.plot([tick, tick+median_width/2], [median_val2, median_val2],
                lw=1, color='k')
    else:
        ax.plot([tick-median_width/2, tick], [median_val1, median_val1],
                lw=1, color='k')
        
        ax.plot([tick, tick+median_width/2], [median_val2, median_val2],
                lw=1, color='k')

legend = ax.legend(loc='lower left')
frame = legend.get_frame()
frame.set_facecolor('white')
frame.set_edgecolor('white')
sns.despine()
fig.tight_layout()

#fig.savefig('crossover.pdf')
```

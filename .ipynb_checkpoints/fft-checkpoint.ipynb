```python
import sys
import cmath
sys.path.append('/home/jjlong/local_gateway/')
from algorithms import np
import json
import seaborn as sns
import matplotlib
%matplotlib inline
import time
def timeit(method):
    def timed(*args, **kw):
        ts = time.time()
        result = method(*args, **kw)
        te = time.time()
        ex_time = te-ts
        return result
    return timed
with open('192.168.123.17.json') as f:
    d = json.load(f)
```

```python
def fft(x):
    N = len(x)
    if N <= 1: return x
    even = fft(x[0::2])
    odd =  fft(x[1::2])
    T= [cmath.exp(-2j*cmath.pi*k/N)*odd[k] for k in range(N//2)]
    return [even[k] + T[k] for k in range(N//2)] + \
           [even[k] - T[k] for k in range(N//2)]
@timeit
def zero_mean(x):
    a = np.Vector(*x)
    return a._zero_mean()
    
```

```python
a = [abs(i) for i in fft(zero_mean(d['x']))]
sns.tsplot(a[0:100])
```
